# EXPLORATORY DATA ANALYSIS

NUMBER OF ATTACKS PER YEAR

```python
attacks_year=dataset["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"attacks_year"}).set_index("Year")
attacks_year.head()
```

```python

attacks_year.plot(kind="bar",color="limegreen",figsize=(15,7))
plt.title("Number of Attacks (Year Wise)")
plt.xlabel("Years")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
#  least number of attacks were made in 1971 and the most number of attacks were made in 2014
```

REGION WISE ATTACKS

```python
pd.crosstab(dataset.Year, dataset.Region).plot(kind='area',stacked = False,figsize=(20,8))
plt.title('Region wise attacks in each Year')
plt.ylabel('Number of Attacks')
plt.xlabel('Year')
plt.show()
```

```python
# most number of attacks took placed in Middle East & North Africa
```

COUNTRY WISE ATTACKS

```python
attack_country = dataset['Country'].value_counts().head(10)
attack_country
```

```python
attack_country.plot(kind = 'bar', color = 'turquoise',figsize = (20,10))
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.title('Country Wise Attacks')
plt.show()
```

```python
# Iraq had most number of attacks took place 
```

PEOPLE KILLED REGION WISE

```python
pkilled_region = dataset[['Region','Kill']].groupby('Region').sum().sort_values(by = 'Kill', ascending = False)
pkilled_region
```

```python
pkilled_region.plot(kind = 'bar', color = 'blueviolet',figsize = (20,10))
plt.xlabel('Region Name')
plt.ylabel('Number of People Killed')
plt.title('People Killed Region Wise')
plt.show()
```

```python
people_Killed = dataset.loc[:,'Kill'].sum()
print(people_Killed)
```

```python
# Total Number of people killed was 411868
```

```python
# Most number of people were killed in Middle East & North Africa and Least number of people were killed in Australasia & Oceanic 
```

PEOPLE WOUNDED (REGION WISE)

```python
pwounded_region = dataset[['Region','Wound']].groupby('Region').sum().sort_values(by = 'Wound', ascending = False)
pwounded_region
```

```python
pwounded_region.plot(kind = 'bar', color = 'blueviolet',figsize = (20,10))
plt.xlabel('Year')
plt.ylabel('Number of People Wounded')
plt.title('People Wounded Region Wise')
plt.show()
```

```python
people_wounded = dataset.loc[:,'Wound'].sum()
print(people_wounded)
```

```python
# Total number of people wounded were 523869
# Most number of people were wounded in Middle East & North Africa
# Least number of people were wounded in Australasia & Oceanic 
```
